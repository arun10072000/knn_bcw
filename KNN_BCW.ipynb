```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
from collections import Counter
import pandas as pd
import random

df = pd.read_csv('breast-cancer-wisconsin.data.txt')
df.replace('?',-99999, inplace=True)
df.drop(['id'], 1, inplace=True)
full_data = df.astype(float).values.tolist()
#print(full_data)

random.shuffle(full_data)
test_size = 0.2
train_set = {2:[], 4:[]} 
test_set = {2:[], 4:[]} 
train_data = full_data[0:-int(test_size*len(full_data))]
test_data = full_data[-int(test_size*len(full_data)):]
                   
for data in train_data:
    train_set[data[-1]].append(data[0:-1])
for data in test_data:
    test_set[data[-1]].append(data[0:-1])
test_set

def knn(data, predict, k=3):
    distances = []
    for group in data:
        for features in data[group]:
            euclidean_distance = np.linalg.norm(np.array(features)-np.array(predict))
            distances.append([euclidean_distance,group])
    votes = [i[1] for i in sorted(distances)[:k]]

    vote_result = Counter(votes).most_common(1)[0][0]
    return vote_result 
correct = 0
total=0
maxk = []
for k in range(1,15,2):    
        for group in test_set:
            for data in test_set[group]:
                vote = knn(train_set, data, k)      
                if group == vote: 
                    correct += 1
                total += 1
        print(k," = ",correct/total) #to get optimal k value
        maxk.append(correct/total)
optimal_k = (maxk.index(max(maxk)))

for group in test_set:
    for data in test_set[group]:
        print(data,end=' ')
        vote = knn(train_set, data, optimal_k)
        print("predction: ",vote)        
        if group == vote: 
            correct += 1
        total += 1
print('Accuracy:',((correct/total)*100),'%')
```
